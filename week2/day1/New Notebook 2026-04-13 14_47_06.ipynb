```sql
CREATE TABLE Employees7 (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);
```

```sql
create table Departments7(
  dept_id INT PRIMARY KEY,
  dept_name VARCHAR(50)
);
```

```sql
create table orders7(
  order_id INT PRIMARY KEY,
  emp_id INT, 
  order_amount INT,
  order_date DATE
);
```

### Find employees who earn more than average salary



```sql
select emp_id, name, salary from employees7
where salary> (select avg(salary) from employees7)
```

### Find employees who belong to the same department as 'Alice'

```sql
select emp_id, name, department_id 
from employees7
where department_id = (select department_id from employees7
where name='Alice'
)
```

### Get employees whose salary is equal to the minimum salary


```sql
select emp_id, name, salary
from employees7 
where salary = (select min(salary)
from employees7)
```

### Find employees who have placed at least one order

```sql
select e.emp_id, e.name , o.order_id
from Employees7 e join Orders7 o
on e.emp_id = o.emp_id
```

### Get employees whose salary is greater than ALL employees in department 102

```sql
select emp_id, name, salary
from employees7
where salary >(select sum(salary)
from employees7 where department_id = 102
)
```

### Create a CTE to show employees with salary > 50,000 and fetch all records

```sql
with my_CTE as (
  select * from employees7
  where salary > 50000
)
select * from my_CTE
```

### Find average salary per department using CTE


```sql
with average_salary as (
  select department_id , avg(salary) as Avg_salary
  from Employees7
  group by (department_id)
)
select * 
from average_salary
```

### Use CTE to get employees and their department names

```sql
with emp_dep_names as (
  select e.name, d.dept_name
  from Employees7 e join Departments7 d 
  on e.department_id = d.dept_id
)
select * 
from emp_dep_names
```
