```sql
-- Replace NULL salary with 0

select emp_id, ifnull(salary, 0) as salary from employees4
```

```sql
-- Replace NULL bonus with 1000
select emp_id, name, ifnull(bonus, 1000) as bonus from employees4
```

```sql
-- Show order amount, if NULL replace with 500

select order_id, ifnull(amount, 500 ) as amount from orders4
```

```sql
-- Replace NULL stock with 0
select product_id, ifnull(stock, 0) as stock from products4
```

```sql
-- Show employee earnings using:
--  salary, if NULL use bonus

select emp_id, name, coalesce(salary, bonus,0) as salary from employees4
```

```sql
-- Show product price:
--  price → 1000 (default)
select product_id, coalesce(price, 1000) from products4
```

```sql
-- Get customer payment:
--  amount → discount → 0
select order_id, coalesce(amount, discount, 0) as amount from orders4
```

```sql
-- Convert salary to NULL if salary = 0

select emp_id,name, Nullif(salary,0) as salary from employees4
```

```sql
-- Convert discount to NULL if discount = 0
select order_id , nullif(discount,0) as discount from orders4
```

```sql
-- Use NULLIF to avoid divide by zero:
-- SELECT amount / NULLIF(discount, 0)
-- FROM Orders;
-- Replace coupon_code with NULL if it is 'DISC10'

select amount/Nullif(discount, 0) as price, nullif(coupon_code,'DISC10')  from orders4
```

```sql
-- Calculate total earnings:
--    salary + bonus (handle NULL properly)
select coalesce(salary,0)+coalesce(bonus,0) as total_earnings from employees4
```

```sql
-- Show employees where:
--    both salary AND bonus are NULL
select emp_id,name from employees4 where salary is NULL and bonus is NULL
```

```sql
-- Show products where:
--    price is NULL but category is NOT NULL

select product_id,product_name from products4 where price is null and category is NOT NULL
```

```sql
-- Show orders where:
--    both amount and discount are NULL

select * from orders4 where amount is null and discount is null 
```

```sql
-- Show employee income:
--   COALESCE(salary, bonus, 1000)

select coalesce(salary,bonus,1000)  as salary from employees4
```

```sql
-- Replace empty discount with NULL using NULLIF:
-- NULLIF(discount, 0)

select order_id, nullif(discount, 0) as discount from orders4
```
