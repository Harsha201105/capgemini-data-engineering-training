```sql
/*16. Find the total sales amount for each employee.
CREATE TABLE Employee (
emp_id INT PRIMARY KEY,
emp_name VARCHAR(50),
department VARCHAR(50),
salary DECIMAL(10, 2),
joining_date DATE
);

CREATE TABLE Sales (
sales_id INT PRIMARY KEY,
emp_id INT,
product VARCHAR(50),
amount DECIMAL(10, 2),
sale_date DATE
);
*/

select e.emp_name as Employee_name, sum(s.amount) as Total_sales
from Employee e inner join Sales s
where e.emp_id=s.emp_id
group by e.emp_name
```

```sql
/*17. List the number of sales made by each employee.*/
select e.emp_name as Employee_name,  count(s.sales_id)
from Employee e inner join Sales s
where e.emp_id = s.emp_id
group by e.emp_name
```

```sql
/*18. Find the total sales amount grouped by product.*/

select product , sum(amount) as sales_amount
from sales
group by product
```

```sql
/*19. Calculate the average sales amount grouped by product.*/
select product, round(avg(amount),2) as Average_Sales_amount
from sales
group by product
```

```sql
/*20. Find employees who have made more than 2 sales, grouped by their names.*/
select e.emp_name as Employee_name
from Employee e inner join sales s
ON e.emp_id = s.emp_id
group by e.emp_name
Having count(s.sales_id)>2
```

```sql
--21. Calculate the total salary and total sales amount for each employee.
select e.emp_name as Employee_name , sum(s.amount) as total_sales
from Employee e inner join Sales s
On e.emp_id = s.emp_id
group by e.emp_name
```

```sql
/*22. Count the number of unique products sold by each employee.*/

select e.emp_name as Employee_name, count( distinct s.product)
from Employee e inner join Sales s
On e.emp_id = s.emp_id
group by e.emp_name
```

```sql
/*23. Find the highest sales amount made by each employee.*/
select e.emp_name, max(s.amount) as Highest_sales_amount
from Employee e inner join Sales s
on e.emp_id = s.emp_id
group by e.emp_name
```

```sql
/* 24 Calculate the total sales amount grouped by product and filtered by products where the total
exceeds 50,000.
*/
select s.product, sum(s.amount) as total_sales_amount
from sales s
group by s.product
Having sum(amount)>50000 
```

```sql
/* 25. Find the departments with the highest average sales amount.*/
select department 
from Employee e inner join Sales s
ON e.emp_id = s.emp_id
group by department
having avg(s.amount)= (select max(avg_salary)
from(
  select avg(s.amount) as avg_salary
  from employee e inner join sales s
  ON e.emp_id = s.emp_id
  group by department
)
)
```

```sql
--26. Find the department with the highest total sales amount.
select e.department
from Employee e inner join Sales s
ON e.emp_id = s.emp_id
group by department
having sum(s.amount) = (select max(total_sales)
from(
  select sum(s.amount) as total_sales
  from Employee e inner join Sales s
  ON e.emp_id = s.emp_id
  group by department

))
```

```sql
-- 27. Show the top 3 employees with the highest total sales amount, grouped by employee
-- names.
select e.emp_name , sum(s.amount) as total_salary
from Employee e inner join Sales s
On e.emp_id = s.emp_id
group by e.emp_name
order by sum(s.amount) desc
limit 3
```

```sql
-- 28. Calculate the total number of employees and the average salary, grouped by the year of
-- joining.
select year(joining_date) as joining_year,count(emp_id) as total_employees, round(avg(salary),2) as average_salary
from employee
group by year(joining_date)
```

```sql
-- 29. Find the total sales amount for each department (using a join between Employee and Sales).

select e.department as Department , sum(s.amount) as total_sales
from Employee e inner join Sales s
On e.emp_id = s.emp_id
group by e.department
```

```sql
-- 30. Show employees who have not made any sales, grouped by their department

select e.emp_name as Employee_name, e.department as department
from Employee e left join Sales s
ON e.emp_id = s.emp_id
where s.sales_id is null
```
