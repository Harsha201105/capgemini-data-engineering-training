Reading the CSV file form catalog


```python
df =spark.read.csv("/Volumes/workspace/default/day1basics/day1dataset.csv",header=True,inferSchema=True)
df.show()
```

```python
df.display()
```

# ### Handle "Blank values" 


```python
from pyspark.sql.functions import *
df=df.replace(["blank","Blank"],None)
df.display()
```

Fix country name

```python
df = df.withColumn("Country", upper(col("Country")))
df.display()
```

Handle missing values


```python
df=df.dropna(subset=["sales","JoinDate","Category"])
df.display()
```

```python
df=df.fillna({"Category":"None",
           "Sales":0,
           "JoinDate":"None"
        })

df.display()
```
